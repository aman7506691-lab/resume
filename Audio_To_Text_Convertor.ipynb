```python
# TO INSTALL THE WHISPER LIBRARY FOR TRANSCRIBE
!pip install openai-whisper
# TO INSTALL ffmpeg FOR AUDIO SUPPORT
!apt install ffmpeg

# TO IMPORT THE WHISPER LIBRARY
import whisper
# TO IMPORT THE FILE HANDLING LIBRARY FROM GOOGLE COLAB
from google.colab import files

# TO SELECT THE MODEL FROM WHISPER
model = whisper.load_model("base")

# TO UPLOAD THE AUDIO FILE FROM SYSTEM
upload_audio = files.upload()
# TO SELECT THE CURRENT AUDIO FROM FILES
current_audio=list(upload_audio.keys())[0]
# TO PASS THE CURRENT AUDIO TO MODEL FOR TRANSCRIBE
output = model.transcribe(current_audio)
# TO PRINT THE TEXT RESULT FROM AUDIO
print(output['text'])
```
